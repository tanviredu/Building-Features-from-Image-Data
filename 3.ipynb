```python
###importing stuff
import skimage
import math
import matplotlib
import numpy             as np
import matplotlib.pyplot as plt
from skimage import io
from skimage import data
from skimage import img_as_float
from skimage import img_as_ubyte

def image_prop(image):
    print("-"*40)
    print("Type :  ",type(image))
    print("DType : ",image.dtype)
    print("Shape : ",image.shape)
    print("size :  ",image.size)
    print("High Pixel  : ",image.max())
    print("Lowes Pixel : ",image.min())
    
    print("-"*40)
    
def plot_image(image):
    plt.figure(figsize=(10,10))
    plt.imshow(image,cmap="gray")
    plt.colorbar()
    image_prop(image)
```

```python
##m reading different channel in a image
image = io.imread("./datasets/images/machu_pichho_peru.jpg")
plot_image(image)
```

```python
## extracting different image
red = image[:,:,0]
green = image[:,:,1]
blue = image[:,:,2]
_,sections = plt.subplots(1,4,figsize=(10,10))
for ax in sections:
    ax.axis('off')
r_section,g_section,b_section,rgb_section = sections
r_section.imshow(red,cmap='gray')
r_section.set_title("Red Channel")
g_section.imshow(green,cmap='gray')
g_section.set_title("Green Channel")
b_section.imshow(blue,cmap='gray')
b_section.set_title("Blue Channel")
rgb_section.imshow(image)
rgb_section.set_title("rgb channel")
```
