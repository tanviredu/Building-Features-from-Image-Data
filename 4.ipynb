```python
###importing stuff
import skimage
import math
import matplotlib
import numpy             as np
import matplotlib.pyplot as plt
from skimage import io
from skimage import data
from skimage import img_as_float
from skimage import img_as_ubyte

def image_prop(image):
    print("-"*40)
    print("Type :  ",type(image))
    print("DType : ",image.dtype)
    print("Shape : ",image.shape)
    print("size :  ",image.size)
    print("High Pixel  : ",image.max())
    print("Lowes Pixel : ",image.min())
    
    print("-"*40)
    
def plot_image(image):
    plt.figure(figsize=(10,10))
    plt.imshow(image,cmap="gray")
    plt.colorbar()
    image_prop(image)
```

```python
## manupulate images colors
image = data.astronaut()
plot_image(image)
```

```python
## we take the red channel and plot the imahe
## this image is a 2D image with 1 channel
## because we extract the channel from the image
blue_ast = image.copy()     ## create a copy of the image
bluish = image[:,:,0] >160  ## take the one channel and then check which pixel of that 2d array is greater then 160
blue_ast[bluish] = [0,0,255]## take all those pixel value from the image and then give them blue color
plot_image(blue_ast)        ## plot the image
```
