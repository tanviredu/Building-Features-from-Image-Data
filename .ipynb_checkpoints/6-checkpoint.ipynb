```python
###importing stuff 
### resizing resclaing cropping and flipping imaegs
### data augmentation
import skimage
import math
import matplotlib
import numpy             as np
import matplotlib.pyplot as plt
from skimage import io
from skimage import data
from skimage import img_as_float
from skimage import img_as_ubyte
from skimage import util
from skimage.color import rgb2gray

def image_prop(image):
    print("-"*50)
    print("Type         :  ",type(image))
    print("DType        :  ",image.dtype)
    print("Shape        :  ",image.shape)
    print("size         :  ",image.size)
    print("High Pixel   :  ",image.max())
    print("Lowes Pixel  :  ",image.min())
    print("Aspect Ratio :  ",image.shape[1]/float(image.shape[0]))
    
    print("-"*50)
    
def plot_image(image):
    plt.figure(figsize=(10,10))
    if image.shape[2] == 3:    
        plt.imshow(image)
    else:
        plt.imshow(image,cmap="gray")
    plt.colorbar()
    image_prop(image)
```

```python
## image transformation. resizing, rescling, rotating
## image augmentation
```

```python
cat = io.imread("./datasets/images/cat.jpg")
plot_image(cat)
```

```python
## resizing the images
## make the image a square images
from skimage.transform import resize
cat_resized = resize(cat,(480,480))
```

```python
plot_image(cat_resized)
```

```python
## resize the image by protecting the aspect ratio of
## of the image
## what is aspect ratio
## aspect ratio is the ratio of the width of the image(shape[1]) and the hight
## of the image (shape[0])

original_cat_asp_ratio = cat.shape[1]/float(cat.shape[0])
resized_cat_asp_ratio = cat_resized.shape[1]/float(cat_resized.shape[0])
print("ORIGINAL ASPECT RATIO IS : {}".format(original_cat_asp_ratio))
print("RESIZED ASPECT RATIO IS  : {}".format(resized_cat_asp_ratio))
```

```python
## when resizing image try to keep the aspect ratio
## otherwise the image will be distorted
## rescale image with the aspect ratio
## we can rescale image with proper aspect ratio
## (1,original_cat_asp_ratio) means height remain same and width will be 1* original_cat_asp_ratio
from skimage.transform import rescale
cat_rescaled = rescale(cat_resized,(1,original_cat_asp_ratio),multichannel=True)
```

```python
plot_image(cat_rescaled)
```

```python
## CROP IMAGE INTHE CENTER
## IN THE CENTER BECAUSE USUALLY MOST OF THE FEATUERS 
## REMAIN IN THE CENTER
```

Imagine you have a big picture, and you want to cut out a smaller square from the center of that picture. The equations you mentioned help you find the exact location where you should start cutting and the exact size of the square you want to cut.

x1 = (W_image - W_crop) / 2:

W_image is the width of your big picture.
W_crop is the width of the square you want to cut out.
We take the width of the big picture, subtract the width of the square you want to cut, and then divide by 2. This tells you how far from the left side you should start cutting to be in the center horizontally.
y1 = (H_image - H_crop) / 2:

H_image is the height of your big picture.
H_crop is the height of the square you want to cut out.
We take the height of the big picture, subtract the height of the square you want to cut, and then divide by 2. This tells you how far from the top you should start cutting to be in the center vertically.
x2 = x1 + W_crop:

This finds the x-coordinate of the right side of your square. We take the starting point x1 and add the width of the square W_crop.
y2 = y1 + H_crop:

This finds the y-coordinate of the bottom side of your square. We take the starting point y1 and add the height of the square H_crop.
When you have these four values (x1, y1, x2, y2), they give you the exact coordinates of the top-left corner (x1, y1) and the bottom-right corner (x2, y2) of the square you want to cut out. Cutting along these coordinates will give you a centered square crop from your big picture.

Think of it like this: You're finding the top-left and bottom-right corners of a window that you're placing over your picture, and this window is your desired square crop. The equations help you position that window right in the center of your picture.

```python
def crop_center(img,cropX,cropY):
    height,width,channels = img.shape
    start_width  = (width//2)  - (cropX//2)
    start_height = (height//2) - (cropY//2) 
    return img[start_height:start_height+cropY,start_width:start_width+cropX]
```

```python
cat_cropped = crop_center(cat_resized,224,224)
```

```python
plot_image(cat_cropped)
```
