```python
###importing stuff
import skimage
import math
import matplotlib
import numpy             as np
import matplotlib.pyplot as plt
from skimage import io
from skimage import data
from skimage import img_as_float
from skimage import img_as_ubyte
from skimage import util
from skimage.color import rgb2gray

def image_prop(image):
    print("-"*40)
    print("Type :  ",type(image))
    print("DType : ",image.dtype)
    print("Shape : ",image.shape)
    print("size :  ",image.size)
    print("High Pixel  : ",image.max())
    print("Lowes Pixel : ",image.min())
    
    print("-"*40)
    
def plot_image(image):
    plt.figure(figsize=(10,10))
    plt.imshow(image,cmap="gray")
    plt.colorbar()
    image_prop(image)
```

```python
## load images
image = data.astronaut()
plot_image(image)
```

```python
### image manupulation
### inverted images
### inversion means origiinal image with min intensity now have maximun intencity
### and vice versa

inverted_image = util.invert(image)
plot_image(inverted_image)
```

```python
## convert the color image to gray
gray_image = rgb2gray(image)
plot_image(gray_image)
```
