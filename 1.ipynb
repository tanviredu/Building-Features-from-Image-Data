```python
###importing stuff
import skimage
import math
import matplotlib
import numpy             as np
import matplotlib.pyplot as plt
from skimage import io
from skimage import data
```

```python
## make a grey scale 2 dimensional image
random_image = np.random.random([500,500])
random_image
```

```python
## display the image
## its a 2 dimensonal image
## represent by 2 array
## image has single channel
## (height,weight,1)
plt.figure(figsize=(10,10))
plt.imshow(random_image,cmap='gray')
plt.colorbar()
```

```python
def image_prop(image):
    print("-"*40)
    print("Type :  ",type(image))
    print("DType : ",image.dtype)
    print("Shape : ",image.shape)
    print("size :  ",image.size)
    print("-"*40)
    
def plot_image(image):
    plt.figure(figsize=(10,10))
    plt.imshow(image,cmap="gray")
    plt.colorbar()
    image_prop(image)
```

```python
## loading a 2 (grayscale) dimensional image
camera = data.camera()
image_prop(camera)
```

```python
plot_image(camera)
```

```python
## loading a 3 dimensional image
chessboard = io.imread("./datasets/images/chessboard.jpg")
plot_image(chessboard)
```

```python
## perform slicing operation 
## perform array indexing 
## since its just a multi dimension aray

## what is the chessboard[:,250:710]
## (:) it means you take all the rows from the image 
## (250:710) means you will take colum from value 0 to 100

plot_image(chessboard[:,0:100])
```

```python
# The expression chessboard[:, 250:510] is a slicing operation on a 2-dimensional array (presumably representing a chessboard) and can be interpreted as follows:

# : in the first position (: before the comma) indicates that you want to select all rows from the array.
# 250:510 in the second position (250:510 after the comma) specifies that you want to select columns starting from index 250 and up to, but not including, index 510.
# So, this slicing operation extracts a portion of the chessboard, including all rows and the columns starting from index 250 up to (but not including) index 510.
```
