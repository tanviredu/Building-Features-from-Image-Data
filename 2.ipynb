```python
###importing stuff
import skimage
import math
import matplotlib
import numpy             as np
import matplotlib.pyplot as plt
from skimage import io
from skimage import data
from skimage import img_as_float
from skimage import img_as_ubyte

def image_prop(image):
    print("-"*40)
    print("Type :  ",type(image))
    print("DType : ",image.dtype)
    print("Shape : ",image.shape)
    print("size :  ",image.size)
    print("High Pixel  : ",image.max())
    print("Lowes Pixel : ",image.min())
    
    print("-"*40)
    
def plot_image(image):
    plt.figure(figsize=(10,10))
    plt.imshow(image,cmap="gray")
    plt.colorbar()
    image_prop(image)
```

```python
## image normalizing . manke 2 sample image
sample1 = np.linspace(0,1,1600).reshape((40,40))
sample2 = np.linspace(1,255,1600).reshape((40,40))
```

```python
plot_image(sample1)
```

```python
plot_image(sample2)
```

```python
## find  the low and high value of the imahe
lowest_value   = sample2.min()
highest_value  = sample2.max()
print("MIN PIXEL VALUE OF IMAGE : {}".format(lowest_value))
print("MAX PIXEL VALUE OF IMAGE : {}".format(highest_value))
```

```python
## normalize the sample image
normalized_sample2_image = sample2/(highest_value-lowest_value)
```

```python
plot_image(normalized_sample2_image)
```

```python
## normalize the image with built in function
coffee = data.coffee()
plot_image(coffee)
```

```python
normalized_image = img_as_float(coffee)
plot_image(normalized_image)
```
